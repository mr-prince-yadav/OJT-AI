```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression

df=pd.read_csv("Mobile phone price.csv")
df

```

```python
df.info()
```

```python
df['Price ($)'] = df['Price ($)'].str.replace(' ', '', regex=False)
df['Price ($)'] = df['Price ($)'].str.replace(',', '', regex=False)
df['Price ($)'] = df['Price ($)'].str.replace('$', '', regex=False)
df['Price ($)']= pd.to_numeric(df['Price ($)'], errors='coerce')
```

```python
model= LogisticRegression(max_iter=10000)
X=df[['Price ($)']]
y=df[['Brand']]
model.fit(X,y.values.ravel())

test_input = pd.DataFrame([[16000]], columns=['Price ($)'])
test=model.predict(test_input)
test
```

```python
import numpy as np
#np.random.seed(19)
sc=np.random.randint(25,100000,11)
sc=np.sort(sc, kind="mergesort")

sc_pred = pd.DataFrame({"Price ($)":sc})

results = model.predict(sc_pred[['Price ($)']])
sc_pred["Predicted Brand"]=results
sc_pred
```
