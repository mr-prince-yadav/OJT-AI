```python
df_2013 = df[df['Order Date'].dt.year == 2013]

# Step 3 & 4: Extract month and group by it
monthly_sales = df_2013.groupby(df_2013['Order Date'].dt.month)['Sales'].sum()

monthly_sales
```

```python
monthly_sales_df = monthly_sales.reset_index()
Month_1 = {1:'January',2:'February',3:'March',4:'April',5:'May',6:'June',7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'}
monthly_sales_df.columns = ['Month', 'Sales']
```

```python
monthly_sales_df['Month']=monthly_sales_df['Month'].map(Month_1)
```

```python
monthly_sales_df
```

```python
plt.figure(figsize=(12,6))
plt.plot(monthly_sales_df["Month"],monthly_sales_df["Sales"])
plt.tight_layout()
```

```python
# Plot
plt.figure(figsize=(12,5))
plt.plot(monthly_sales_df["Month"],monthly_sales_df["Sales"], color="green", linestyle="-.", marker="*", markersize="10", linewidth=2, label="Sales data")
plt.fill_between(monthly_sales_df["Month"],monthly_sales_df["Sales"], color='green', alpha=0.1)

# Plot Labels
plt.xlabel("Month",fontsize=14)
plt.ylabel("Sales",fontsize=14)
plt.title("Monthly Sales Data for 2025", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.legend()
plt.show()
```
