```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np

months = [ 
    "2025-01", "2025-02", "2025-03", "2025-04", "2025-05", "2025-06",
    "2025-07", "2025-08", "2025-09", "2025-10", "2025-11", "2025-12"
]
sales = [
    67850.30, 63240.75, 71450.90, 76890.25, 43210.50, 82345.60,
    89760.35, 86430.20, 82150.75, 91240.30, 103560.45, 118790.80
]

# Convert months to datetime objects for better plotting
months = [mdates.datestr2num(m + "-01") for m in months]

plt.figure(figsize=(12,6))

# Plotting the line with additional style
plt.plot(months, sales, color='green', linestyle='--', marker='*', markersize=10, label="Sales Data")

# Filling the area under the curve with a gradient color
plt.fill_between(months, sales, color='green', alpha=0.1)

# Formatting the x-axis with month names
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b-%Y'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gcf().autofmt_xdate()

# Adding titles and labels
plt.title("Monthly Sales Data for 2025", fontsize=16, fontweight='bold')
plt.xlabel("Month", fontsize=14)
plt.ylabel("Sales ($)", fontsize=14)

# Adding a grid and improving visibility
plt.grid(True, linestyle='--', alpha=0.5)

# Adding annotations
plt.annotate('Highest sales in Dec', xy=(months[11], sales[11]), xytext=(months[10], sales[11] + 20000),
             arrowprops=dict(facecolor='red', shrink=0.05), fontsize=12)

# Show legend
plt.legend()

plt.tight_layout()
plt.show()
```

# Double Line Plot

```python
year=[1972,1982,1992,2002,2012]
e_india=[100.6,158.61,305.54,394.96,724.79]
de_nepal=[10.5,25.21,58.65,119.27,274.87]

plt.figure(figsize=(10,5))
plt.plot(year,e_india, color="green", linestyle="-", marker="o", markersize="5", linewidth=2, label="India")
plt.plot(year,de_nepal, color="orange", linestyle="--", marker="o", markersize="5", linewidth=2, label="Nepal")

plt.fill_between(year,e_india, color='green', alpha=0.1)
plt.fill_between(year,de_nepal, color='orange', alpha=0.1)

# Plot Labels

plt.title("Energy Consumption: India vs Nepal (1972-2012)", fontsize=16, fontweight='bold')
plt.xlabel("Year",fontsize=14)
plt.ylabel("Energy Consumption (in TWh)",fontsize=14)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show
```
