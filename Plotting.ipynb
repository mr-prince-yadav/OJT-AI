```python
# Plotting Two basic varibles
import matplotlib.pyplot as plt

x=[5,2,9,4,7]
y=[10,5,8,4,2]

plt.plot(x,y)
plt.title("PLot")
plt.show
```

```python
plt.title("Bar")
plt.bar(x,y)
plt.show
```

```python
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.show
```

```python
plt.pie(x, labels=["A","B","C","D","E"],colors=["Red","Yellow","Brown","Blue","Green"], autopct='%1.1f%%')
plt.title("Pie chart")
plt.axis('equal')
plt.show()
```
